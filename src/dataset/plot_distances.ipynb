```python
"""This Notebook requires the .npy files that contains the distances of the object to the camera. 
To generate these .npy files use the statistics.py to calculate these values for the current dataset. 
"""

import os
import sys

import numpy as np

sys.path.append(os.path.abspath("../"))

from pathlib import Path

import matplotlib.pyplot as plt
```

```python
path_to_distances = Path("..", "..", "data", "distances")
```

```python
distances = np.load(Path(path_to_distances, "ball.npy"))

# Define the bin edges (e.g., bins of width 1)
bin_edges = np.arange(0, 13 + 2, 0.1)  # Adjust as needed

counts, bin_edges = np.histogram(distances, bins=bin_edges)
plt.figure(figsize=(10, 6))
plt.bar(
    bin_edges[:-1],  # Use the left edge of each bin for the x-axis
    counts,  # Use the counts for the y-axis
    width=np.diff(bin_edges),  # Set the width of each bar
    color="skyblue",
    edgecolor="black",
    align="edge",  # Align bars with the bin edges
)

# Add labels and title
plt.xlabel("Distance")
plt.ylabel("Frequency")
plt.title("Distribution of Distances for Balls")
plt.xticks(np.arange(0, bin_edges[-1], 1))

plt.show()
```
